```python
# modules we'll use
import pandas as pd
import numpy as np

# read in all our data
data1 = pd.read_csv("data/NFL Play by Play 2009-2017 (v4).csv")
data2 = pd.read_csv("data/Building_Permits.csv")

# set seed for reproducibility
np.random.seed(0) 
```

```python
data1.sample(5)
```

Now we can see how many missing data points we have on the nfl data

```python
# get the number of missing data points per column
missing_values_count_data1 = data1.isnull().sum()

# look at the # of missing points in the first ten columns
missing_values_count_data1[0:10]
```

```python
# how many total missing values do we have?
total_cells_data1 = np.product(data1.shape)
total_missing_data1 = missing_values_count_data1.sum()

# percent of data that is missing
(total_missing_data1/total_cells_data1) * 100
```

Removing all the missing values rom the nfl data

```python
# remove all the rows that contain a missing value
data1.dropna()
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped_data1 = data1.dropna(axis=1)
columns_with_na_dropped_data1.head()
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % data1.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped_data1.shape[1])
```

Filling in missing values automatically

```python
# get a small subset of the NFL dataset
data1 = data1.loc[:, 'EPA':'Season'].head()
data1
```

```python
# replace all NA's with 0
data1.fillna(0)
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the reamining na's with 0
data1.fillna(method = 'bfill', axis=0).fillna(0)
```
